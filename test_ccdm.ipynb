```python
Ak = np.zeros([2,4])
v0 = np.array([[1], [8]])
vk = v0
# 构建星座平方块矩阵
blockreal = np.array([1,2,3,4])
blockimag = np.array([4,3,2,1])

Ak[0] = blockreal
Ak[1] = blockimag
Ck = np.matmul(Ak, Ak.transpose())
print("Ck: " + str(Ck))
vk = np.matmul(Ck, vk)
vk = vk / np.sqrt(np.sum(vk ** 2))
print("vk: " + str(vk))
phiK = 0.5 * np.arctan2(vk[0][0], vk[1][0]) - (np.pi / 4)
print(phiK)
```
